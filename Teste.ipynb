```python
import pandas as pd

df = pd.read_csv(r'netflix.csv')
```

```python
df.info()
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], format = '%m/%d/%Y', errors = 'coerce')
```

```python
df['release_year'] = pd.to_datetime(df['release_year'], format = '%Y', errors = 'coerce')
```

```python
# shape, tipos, nulls, duplicados.
def info_gerais(df):
  print(f'QTD Linhas Dataset: {df.shape[0]}')
  print(f'QTD Colunas Dataset: {df.shape[1]}\n')

  for i in df.columns:
    print(f'Coluna {i}\n')
    print(f'QTD Nulos: {df[i].isnull().sum()}')
    print(f'% Nulos: {df[i].isnull().sum()/len(df)}\n')
    print(f'QTD Duplicados: {df[i].duplicated().sum()}\n')
    print(f'Quantidade por categoria: {df[i].value_counts()}\n')
    print(f'Quantidade por categoria: {df[i].value_counts()/len(df)*100}\n')
    print(f'Valores distintos: {df[i].nunique()}\n')
```

```python
# Vê todos os títulos duplicados
df[df['title'].duplicated(keep=False)].sort_values('title')
```

```python
info_gerais(df)
#drop director
#
```

```python
df[df['title'].duplicated(keep = False )]
```

```python
df.drop_duplicates(subset = ['title'], keep= 'last', inplace = True)
```

```python
df.drop(columns = ['director'], inplace = True)
```

```python
df[['Listed 1', 'Listed 2', 'Listed 3']] = (
    df['listed_in']
    .str.split(',', expand=True)
    .iloc[:, :3]
)
```

```python
df['rating'] = df['rating'].replace({'UR': 'NR'})
```

```python
# replace — só substitui o que está no dicionário, mantém o resto
df['Métrica'] = df['Métrica'].replace({'Seasons': 'Season'})
```
